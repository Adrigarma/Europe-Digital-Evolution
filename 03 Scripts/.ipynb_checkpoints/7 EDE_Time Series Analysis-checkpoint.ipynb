# 5. Testing for stationarity

```python
from statsmodels.tsa.stattools import adfuller # import function

def dickey_fuller(timeseries): # define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic', 'p-value', 'Number of Lags Used', 'Number of observations Used'])
    for key, value in test[4].items():
        result['Critical Value (%s)' %key] = value
    print(result)

# Apply the test using the function on the time series
dickey_fuller(trend_all['internet_access_rate'])
```

- p-value not significant
- Test statistic is larger than the critical values
- The null hypothesis can't be disproved.
- **The data is non-stationary**

# 6. Testing for autocorrelations

```python
# Check out a plot of autocorrelation

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(trend_all)
plt.show()
```

- First lags are correlated
- **The data is non-stationary**

# 7. Stationarizing the Data

```python
data_diff = trend_all - trend_all.shift(1) # the df.shift(1) function turns the observation to t-1
```

```python
data_diff.dropna(inplace=True) # remove the missing values that came as a result of the differencing. 
```

```python
data_diff.head()
```

```python
data_diff.columns
```

```python
# checking out waht the differencing did to the time series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

```python
dickey_fuller(data_diff)
```

- p-value still non-significant
- Test Statistics still larger than critical values
- **The data is still non-stationary**

```python
plot_acf(data_diff)
plt.show()
```

- All lags are inside the confidence band

```python
# Second differencing 
data_diff_2 = data_diff - data_diff.shift(1) # the df.shift(1) function turns the observation to t-1
```

```python
data_diff_2.dropna(inplace=True) # remove the missing values that came as a result of the differencing.
```

```python
data_diff_2.head()
```

```python
# check curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff_2)
```

```python
dickey_fuller(data_diff_2)
```

- p-value is significant (0.001) below 0.05, so the null hypothesis can be rejected. 
- Test Statistics is smaller than critical values at 1%, 5% and 10% confirming the stationarity. 
- **The data is stationary**

```python
plot_acf(data_diff_2)
plt.show()
```

- All lags are inside the confidence are confirming the stationarity of the data.

There are two lag otiside the blue area, so I will use p=1

The plot shows  3 lags outside the confidence area, so I will use q=1

## 8.2 Splitting the Data

```python
trend_all.shape
```

```python
trend_all
```

```python
# Split the data

train = trend_all.loc[:'2018-01-01']
test = trend_all.loc['2019-01-01':]
```

```python
# Plot both sets in order to see the split

plt.figure(figsize=(15,5), dpi=100)
plt.plot(trend_all, color='lightgray')
plt.plot(train)
plt.plot(test)
```

```python
test.shape
```

## 8.3 Running and Fitting the Model

```python
# Importing model to forecasting 

from statsmodels.tsa.api import ARIMA # This case is ARIMA becasue the data was differenced 2 times
```

```python
model = ARIMA (train, order=(1,2,1))
fitted = model.fit()
print(fitted.summary()) 

# Forecast
steps = len(test)
fc_obj = fitted.get_forecast(steps=steps)
fc = fc_obj.predicted_mean
conf = fc_obj.conf_int(alpha=0.05) # 95% confidence interval

# Align index with test
fc.index = test.index
conf.index = test.index

# Make pandas series
fc_series = pd.Series(fc,index=test.index, name='forecast')
lower_series = conf.iloc[:, 0]
upper_series = conf.iloc[:, 1]

# Plot
plt.figure(figsize=(12,4), dpi=100)
plt.plot (train, label='training')
plt.plot (test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(test.index, lower_series, upper_series,
                 color='k', alpha=.05) # This creates the confidence interval
plt.title('Forecast vs Actual')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

- The AR coefficient is statistically significant (p=0.003), wihle the MA coefficinet is not (p=0.241)
- The forecast line overlap the actual line and remains within the coefficient interval.
- I will re-estimate the model without the MA parameter, as it does not appear to contribute to the predictions. 

```python
# Second model estimate

model = ARIMA (train, order=(1,2,0))
fitted = model.fit()
print(fitted.summary()) 

# Forecast
steps = len(test)
fc_obj = fitted.get_forecast(steps=steps)
fc = fc_obj.predicted_mean
conf = fc_obj.conf_int(alpha=0.05) # 95% confidence interval

# Align index with test
fc.index = test.index
conf.index = test.index

# Make pandas series
fc_series = pd.Series(fc,index=test.index, name='forecast')
lower_series = conf.iloc[:, 0]
upper_series = conf.iloc[:, 1]

# Plot
plt.figure(figsize=(12,4), dpi=100)
plt.plot (train, label='training')
plt.plot (test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(test.index, lower_series, upper_series,
                 color='k', alpha=.05) # This creates the confidence interval
plt.title('Forecast vs Actual')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

- The coefficient is 0.000 which is significant, but the forecast line is further as the past model from the actual line.
- I will try now with p=2 

```python
# Third model estimate

model = ARIMA (train, order=(2,2,0))
fitted = model.fit()
print(fitted.summary()) 

# Forecast
steps = len(test)
fc_obj = fitted.get_forecast(steps=steps)
fc = fc_obj.predicted_mean
conf = fc_obj.conf_int(alpha=0.05) # 95% confidence interval

# Align index with test
fc.index = test.index
conf.index = test.index

# Make pandas series
fc_series = pd.Series(fc,index=test.index, name='forecast')
lower_series = conf.iloc[:, 0]
upper_series = conf.iloc[:, 1]

# Plot
plt.figure(figsize=(12,4), dpi=100)
plt.plot (train, label='training')
plt.plot (test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(test.index, lower_series, upper_series,
                 color='k', alpha=.05) # This creates the confidence interval
plt.title('Forecast vs Actual')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

- This model shows coefficients of 0.000 which make it statistically significant.
- The forecast line overlaps with the actual data.
- **This is a good fit**
