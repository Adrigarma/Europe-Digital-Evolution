## 3.3 Deriving New Variables

### There are too many different activities that could make the analysis very complex, so I decided to make 9 groups of activities: 
### 1. Personal Communication
- sending/receiving e-mails  
- telephoning or video calls

### 2. Information and Research
- finding information about goods and services  
- consulting wikis (to obtain knowledge on any subject)  
- reading/downloading online newspapers/news

### 3. E-commerce and Services
- Internet banking  
- selling goods or services  
- travel and accommodation services  

### 4. Social Media 
- participating in social networks (creating user profile, posting messages or other contributions to facebook, twitter, etc.)  
- uploading self-created content to any website to be shared  
- sharing or publishing self-created videos, photos, music, texts, etc on a website or via an app  
- posting messages to social media sites or instant messaging

### 5. Learning and Education
- doing an online course (of any subject)  
- communicating with instructors or students using educational websites/portals  

### 6. Work and Professional Use
- job search or sending an application  
- participating in professional networks (creating user profile, posting messages or other contributions to LinkedIn, Xing, etc.)
- creating websites or blogs

### 7. Media and Entertainment
- watching internet streamed TV or videos  
- playing/downloading games, images, films or music  
- listening to web radios and/or watching web TV  
- downloading software

### 8. Health
- accessing personal health records online  
- making an appointment with a practitioner via a website

### 9. Civic
- civic or political participation

```python
# Deriving new variable
def activity_group(row):
    if row['internet_activities'] in ['sending/receiving e-mails', 
                                      'telephoning or video calls']:
        return 'Personal Communication'
        
    elif row['internet_activities'] in ['consulting wikis (to obtain knowledge on any subject)', 
                                        'finding information about goods and services', 
                                        'reading/downloading online newspapers/news']:
        return 'Information and Research'
        
    elif row['internet_activities'] in [ 'Internet banking', 
                                        'selling goods or services', 
                                        'travel and accommodation services',
                                        'managing a payment account to (potentially) pay for goods or services purchased over the internet']:
        return 'E-commerce and Services'
        
    elif row['internet_activities'] in ['participating in social networks (creating user profile, posting messages or other contributions to facebook, twitter, etc.)',
                                         'uploading self-created content to any website to be shared', 
                                         'sharing or publishing self-created videos, photos, music, texts, etc on a website or via an app',
                                         'posting messages to social media sites or instant messaging']:
         return 'Social Media'
         
    elif row['internet_activities'] in [ 'doing an online course (of any subject)',
                                        'communicating with instructors or students using educational websites/portals']: 
        return 'Learning and Education'

    elif row['internet_activities'] in ['job search or sending an application',
                                        'participating in professional networks (creating user profile, posting messages or other contributions to LinkedIn, Xing, etc.)',
                                        'creating websites or blogs']:
        return 'Work and Professional Use'

    elif row['internet_activities'] in [ 'watching internet streamed TV or videos',
                                        'playing/downloading games, images, films or music',
                                        'listening to web radios and/or watching web TV',
                                        'downloading software']:
        return 'Media and Entertainment'

    elif row['internet_activities'] in ['accessing personal health records online',
                                        'making an appointment with a practitioner via a website']:
        return 'Health'

    elif row['internet_activities'] in ['civic or political participation']:
        return 'Civic'

    else:
        return 'Other'
```

```python
# Apply function
df['activity_group'] = df.apply(activity_group, axis=1)
```

```python
df['activity_group'].value_counts()
```

```python
df[df['activity_group'] == 'Other']
```

```python
# Verifying that all 'Other' activity_group correspond to NaN values in internet_activities
df[df['activity_group'] == 'Other']['internet_activities'].isna().sum()
```

The 'Other' activities are NaN values for internet activities. These NaN values were generated when the merging of datasets was done. So they will not be included in the analysis when analyzing internet activities. 

## 3.4 Correlations:

Due to the nature of the dataset, which includes both categorical and numerical values, a straightforward correlation analysis was not possible. In this case, the numerical values are directly associated with categorical values. Therefore, the numeric variables cannot be analyzed in isolation, as their meaning depends on the categories they are linked to. 
A thoughtful wrangling process was required to restructure the data in a way that would make a correlation heatmap meaningful and interpretable. 
After some exploration, the most meaningful correlation analysis was found to be between internet activity groups and purchasing power per capita, aggregated by country. 
To achieve this I did the next steps:
1. I filtered the dataset to include only the demographic group "All individuals".
2. Then, I grouped the data by country and activity group, calculating the mean internet activity rate for each combination.
3. In order to create a matrix suitable for the analysis, I created a pivot table, so that each "activity_group" became a column, with countries as rows.
4. Finally, I merged the purchasing_power_gdp per country into the pivot table previously created to include it in the analysis.

The resulting dataset allowed me to generate the correlation heatmap that shows the relationship between different types of internet activities and the purchasing power indicator across countrires. 

The different steps are shown below: 

```python
# Create a subset for demographic_group = All individuals
df_all = df[df['demographic_group'] == 'All individuals'].copy()
df_all
```

```python
# Group by country and activity_group
df_group = df_all.groupby(['country', 'activity_group'])['internet_activities_rate'].mean().reset_index()
```

```python
df_group
```

```python
# Create pivot to make each activity_group a column
pivot_act = df_group.pivot(index = 'country', columns = 'activity_group', values = 'internet_activities_rate')
pivot_act.head()
```

```python
# Delete 'Other' column
pivot_act = pivot_act.drop(columns='Other')
pivot_act.columns
```

```python
# Include purchasing power to pivot table
pivot_act['purchasing_power_gdp'] = df_all.groupby('country')['purchasing_power_gdp'].mean()
```

```python
pivot_act.head()
```

```python
pivot_act.describe()
```

```python
# Correlation table
pivot_act.corr()
```

```python
# Create a subplot with matplotlib
f, ax = plt.subplots(figsize = (8,8))

# Create the correlation heatmap
corr_act_gdp = sns.heatmap(pivot_act.corr(), annot = True, ax = ax)
```

```python
# Save graph
f.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'corr_heatmap_activities_gdp.png'))
```

### Insights:
As shown in the graph, there is a strong positive correlation between certain activity groups, such as: E-Commerce and Services, Information and Research, Learning and Education, Personal Communication and Work and Professional Use and purchasing power (GDP per capita). This suggests that using the internet for these types of activities may be associated with higher levels of purchasing power. 

On the other hand, activities such as Civic, Health, Media and Entertainment and Social Media show a weaker correlation with purchasing power, indicating that they may be more equally accessed regardless of the economic purchasing power.  

Personal Communication activities are strongly correlated with almost all other activity groups, except for Health. This may suggest a lag in the digitalization within the healthcare sector. 

Activities related to Work and Professional Use also show very strong correlations with E-commerce and Services, Information and Resrarch, Learning and Education and Personal Communication. These could suggest tha people who use digital tools for work are also more likely to engage in other online services, such as online banking, e-commerce, learning platforms and information seeking.

## 3.7 Categorical Plots:

A categorical plot will ve created to split the 'purchasing_power_gdp' column into categories.

```python
# Visualize a histogram to see the distribution of variables to determine categories. 
hist, ax = plt.subplots()
sns.histplot(pivot_act['purchasing_power_gdp'], bins = 20, kde = True)
plt.show()
```

```python
# Save graph
hist.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'gdp_country_hist.png'))
```

### Insights:
The histogram shows a right-skewed distribution of purchasing power GDP across EU member countries. Most countries fall between 10,000 and 40,000, while only a few (around 4) exceed 40,000. 

Based on this distribution, three categories will be created:
- Low GDP: < 20,000
- Medium GDP: 20,000-40,000
- High gdp: > 40,000

```python
pivot_act.loc[pivot_act['purchasing_power_gdp'] < 20000, 'GDP Category'] = 'Low GDP'
```

```python
pivot_act.loc[(pivot_act['purchasing_power_gdp'] >= 20000) & (pivot_act['purchasing_power_gdp'] < 40000), 'GDP Category'] = 'Middle GDP'
```

```python
pivot_act.loc[pivot_act['purchasing_power_gdp'] >= 40000, 'GDP Category'] = 'High GDP'
```

# 4. Hypothesis

Based on the exploratory analysis I get to the next hypothesis:
