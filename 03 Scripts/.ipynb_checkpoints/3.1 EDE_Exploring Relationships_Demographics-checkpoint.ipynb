# 3. Exploring Relationships 

```python
# Simplify demographic groups in new variable
def demographic_group_category(row):
    if row['demographic_group'] in ['Females, 16 to 24 years old', 
                                    'Females, 25 to 34 years old',
                                    'Females 35 to 44 years old', 
                                    'Females 45 to 54 years old',
                                    'Females 55 to 64 years old', 
                                    'Females 65 to 74 years old',
                                    'Males, 16 to 24 years old',
                                    'Males, 25 to 34 years old', 
                                    'Males 35 to 44 years old',
                                    'Males 45 to 54 years old', 
                                    'Males 55 to 64 years old',
                                    'Males 65 to 74 years old']:
        return 'Gender & Age'
    elif row['demographic_group'] in ['Females with low formal education', 
                                      'Females with medium formal education', 
                                      'Females with high formal education',
                                      'Males with low formal education',
                                      'Males with medium formal education',
                                      'Males with high formal education']:
        return 'Gender & Education level'
    elif row['demographic_group'] in ['Employees working full time', 
                                      'Employees working part time',
                                      'Retired Individuals',
                                      'Self-employed, family workers', 
                                      'Students',
                                      'Unemployed']:
        return 'Employment Status'
    elif row['demographic_group'] in ['Individual living in a household with income in first quartile',
                                      'Individual living in a household with income in second quartile',
                                      'Individual living in a household with income in third quartile',
                                      'Individual living in a household with income in fourth quartile']:
        return 'Income quartiles'
    elif row['demographic_group'] in ['Individuals living in cities',
                                      'Individuals living in towns and suburbs',
                                      'Individuals living in rural areas']:
        return 'Area of Residence'
    elif row['demographic_group'] in ['Individuals living in a household with children',
                                  'Individuals living in a household without children']:
        return 'Household status'
    elif row['demographic_group'] in ['ICT professionals', 
                                  'Non ICT professionals',
                                  'Mobile internet users', 
                                  'Non-users of mobile internet']:
        return 'Special interests'
    else: 
        return 'Other'
```

```python
# Apply function
df['demographic_group_category'] = df.apply(demographic_group_category, axis=1)
```

```python
df['demographic_group_category'].value_counts()
```

```python
# Export dataframe with demographic_category
df.to_csv(os.path.join(path, '02 Data', 'Prepared Data','internet_use_act_pp_gdp_act_group_demo_cat.csv'))
```

The demographic bar chart shows that the top 5 activity groups are consistent across all age groups, with Informaton and Research being the most dominant activity group. This aligns with the correlation heatmap, where Information and Research has a strong connection with both E-commerce & Services and Media & Entertainment, which are also within the top activities.  
Gender differences appear minimal, while age has a much stronger effect on the internet activity rates. Younger groups shows more engagement across activities.

## 4. Conclusions & Hypothesis

The bar chart analysis shows that across most demographic segments (age, education, residence), the hierarchy of activities is stable: Information & Research, Personal Communication, Social Media, and Media & Entertainment consistently appear at the top, with differences mainly in intensity (higher among younger, highly educated, and urban groups; lower among older, less educated, and rural groups). Employment, however, not only affects the overall levels but also changes the shape of the activity profile, with students standing out in Learning & Education, unemployed individuals in Work & Professional Use, retirees showing very low participation, and employees maintaining a balanced profile.

The heatmap correlations confirm this pattern: Information & Research is systematically the most central activity, strongly connected with Media & Entertainment, E-commerce & Services, and Personal Communication. By contrast, Work & Professional Use consistently appears as the least integrated activity, forming a separate and more isolated dimension of internet use across all groups.

Based on this exploratory analysis, we derive the following **hypotheses**:

- H1: Age, gender, education level and area of residence affect the intensity of internet use, but do not change the hierarchy of activities. 
- H2: Employment status affects both intensity and hierarchy of internet use for different activities. 
- H3: Information & Research is associated with Media & Entertainment, E-Commerce and Personal Communication, while Work & Professional Use remains least integrated activity across all demographic groups.
